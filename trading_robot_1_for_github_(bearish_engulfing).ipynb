```python
#add candle pattern for strategy
def find_bearish_engulfing(newdf):
  bearish_engulfing_list = []
  for i in range(len(newdf)):
    if i < (len(newdf))-5:
      #bearish patterns
      #bearish engulfing
      if newdf['open'][i+1] < newdf['close'][i+1] and newdf['open'][i] > newdf['close'][i] and newdf['close'][i] < newdf['close'][i+1] and abs(newdf['close'][i] - newdf['open'][i]) > abs(newdf['close'][i+1] - newdf['open'][i+1]) and newdf['high'][i] > (newdf['high'][i+1]+newdf['high'][i+2]+newdf['high'][i+3]+newdf['high'][i+4]+newdf['high'][i+5])/5:
        bearish_engulfing_list.append(i)
  return(bearish_engulfing_list)

#add bearish engulfing
euraud_1day['bearish_engulfing'] = 0
for i in find_bearish_engulfing(euraud_1day):
  euraud_1day['bearish_engulfing'][i] = 1

chfjpy_1day['bearish_engulfing'] = 0
for i in find_bearish_engulfing(chfjpy_1day):
  chfjpy_1day['bearish_engulfing'][i] = 1
```

```python
#create statistics function
def statistics(newdf, points_list):
  gains=0
  ng=0
  losses=0
  nl=0
  totalR=0
  drawdown=0
  drawdown_list = []

  for i in range(len(points_list)):
    if newdf.values[0][0]//10 > 5 and max(max(points_list), abs(min(points_list))) > 500:
      points_list[i] = points_list[i]/100
    elif newdf.values[0][0]//10 == 1 and max(max(points_list), abs(min(points_list))) > 500:
      points_list[i] = points_list[i]/10

  for i in points_list:
    if(i>0):
      gains+=i
      ng+=1
    else:
      losses+=i
      nl+=1
    totalR += i

  if(ng>0):
    avgGain=int(gains/ng)
    maxR=str(int(max(points_list)))
  else:
    avgGain=0
    maxR="undefined"

  if(nl>0):
    avgLoss=int(losses/nl)
    maxL=str(int(min(points_list)))
    ratio=str(-avgGain/avgLoss)
  else:
    avgLoss=0
    maxL="undefined"
    ratio="inf"

  if(ng>0 or nl>0):
    battingAvg=ng/(ng+nl)
  else:
    battingAvg=0

  for i, j in enumerate(points_list):
    if i < (len(points_list)-1) and j < 0:
      drawdown += j
    elif i < (len(points_list)-1) and j > 0:
      drawdown_list.append(drawdown)
      drawdown = 0
    elif i == (len(points_list)-1) and j < 0:
      drawdown += j
      drawdown_list.append(drawdown)
    elif i == (len(points_list)-1) and j > 0:
      drawdown_list.append(int(round(drawdown,0)))
  try:
    max_drawdown = int(min(drawdown_list))
  except:
    max_drawdown = 0
  profit = int(totalR)

  print()
  print("Results for " +" going back to "+str(newdf.index[-1])+", Sample size: "+str(ng+nl)+" trades")
  print("Batting Avg: "+ str(battingAvg))
  print("Gain/loss ratio: "+ ratio)
  print("Average Gain: "+ str(avgGain))
  print("Average Loss: "+ str(avgLoss))
  print("Max Return: "+ maxR)
  print("Max Loss: "+ maxL)
  print("Total return over "+str(ng+nl)+ " trades: "+ str(profit)+" points" )
  print("Max drawdown is " + str(max_drawdown) + " points")
  print()
  return(profit, max_drawdown)
```

```python
#strategy 1 (bearish_engulfing)
def bearish_engulfing(newdf, tp, sl, ep):
  df = newdf.copy()
  df = df.iloc[::-1]
  pos_exec = 0 #the flag if we entered the market
  pos_exist = 0 #the flag for check if exist the position
  sell_limit_exist = 0 #the flag if we put the sell limit
  points_list = []
  datetime_list = []
  num = 0
  sell_limit_date = dt.datetime.today()
  
  for i in range(len(df)):
    if sell_limit_exist == 0 and pos_exist == 0 and df['bearish_engulfing'][i] == 1:
      datetime = []
      sell_limit = df['open'][i+1] + ep
      print('sell limit added after price closed at ' + str(round(df.close[i],6)) + ' ' + str(df.index[i].date()))
      sell_limit_exist = 1
      sell_limit_date = df.index[i].date()
      datetime.append(df.index[i].date())
    #delete sell limit after 2 weeks without order
    elif pos_exist == 0 and sell_limit_exist == 1 and (df.index[i].date() - sell_limit_date).total_seconds()/86400 > 14:
      print('The order was not activated')
      sell_limit_exist = 0
      pos_exist = 0
    elif pos_exist == 0 and sell_limit_exist == 1 and df['high'][i] >= sell_limit:
      sell_price = sell_limit
      print('sell now from ' + str(round(sell_price,5)) + ' ' + str(df.index[i].date()) + ' at ' + str(df.index[i].time()))
      pos_exec = 1
      pos_exist = 1
      sell_limit_exist = 0
      datetime.append(df.index[i].date())
    elif pos_exec == 1 and df['high'][i] >= (sell_price + sl):
      pos_exec = 0
      pos_exist = 0
      close_price = (sell_price + sl)
      print('close position now at ' + str(round(close_price,5)) + ' ' + str(df.index[i].date()) + ' at ' + str(df.index[i].time()) + ' stop loss :(')
      datetime.append(df.index[i].date())
      points = int((sell_price - close_price)*10000)
      points_list.append(points)
      datetime_list.append(datetime)
    elif pos_exec == 1 and df['low'][i] <= (sell_price - tp):
      pos_exec = 0
      pos_exist = 0
      close_price = (sell_price - tp)
      print('close position now at ' + str(round(close_price,5)) + ' ' + str(df.index[i].date()) + ' at ' + str(df.index[i].time()) + ' take profit :)')
      datetime.append(df.index[i].date())
      points = int((sell_price - close_price)*10000)
      points_list.append(points)
      datetime_list.append(datetime)
    elif(num==df["close"].count()-1 and pos_exec==1):
      pos_exec=0
      close_price = df['close'][i]
      print("Selling now at "+ str(close_price))
      datetime.append(df.index[i].date())
      points = int((sell_price - close_price)*10000)
      points_list.append(points)
      datetime_list.append(datetime)
    
    num += 1
      
  return(points_list, datetime_list)
```
